```python
import numpy as np
import matplotlib.pyplot as plt

def plot_generated_log(start,end,num):
    # Generate values using logspace with non-uniform spacing
    values = np.logspace(np.log10(start), np.log10(end), num=num)
    print(values)
    
    # Round the values to integers
    rounded_values = np.round(values).astype(int)
    print(rounded_values)
    
    # Plot the values
    plt.figure(figsize=(8, 6))
    plt.plot(values, marker='o', linestyle='-')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title('Values Generated Using Logarithmic Scale')
    plt.grid(True)
    plt.show()

plot_generated_log(0.001,1,10)
```

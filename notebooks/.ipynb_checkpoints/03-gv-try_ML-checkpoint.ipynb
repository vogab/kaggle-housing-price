```python
#Save the 'Id' column
train_ID = train['Id']
test_ID = test['Id']

# remove Id col for further preprocessing
train_noid=train.drop('Id',axis=1)
test_noid=test.drop('Id',axis=1)

print(train_noid.shape)
print(test_noid.shape)
```

## 2.1. xgboost

### clean fine tuning

```python
# the grid search for best params. beware, time consuming to run

import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error





# Define the model
model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'alpha': [0, 0.1, 1],
    'lambda': [1, 1.5, 2]
}

# Perform grid search
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Best parameters
print("Best parameters found: ", grid_search.best_params_)
print("Best score: ", -grid_search.best_score_)

# Train model with best parameters
best_params = grid_search.best_params_
best_model = xgb.XGBRegressor(**best_params, objective='reg:squarederror', random_state=42)
best_model.fit(X_train, y_train)

```

```python
import json
import pickle


# save model and params

# Save best params to a JSON file
with open('best_params_xgboost.json', 'w') as f:
    json.dump(best_params, f)

with open('grid_search_xgboost.pkl', 'wb') as f:
    pickle.dump(grid_search, f)
```

```python
# Evaluate the model
y_pred = best_model.predict(X_val)
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
print("Validation RMSE: ", rmse)
```

## 2.2. lasso regression

### base model out of the box

```python
from sklearn.linear_model import Lasso


# Instantiate and train the Lasso regression model
lasso = Lasso(alpha=0.0005, random_state=42)
lasso.fit(X_train, y_train)

# Make predictions on the training set
y_train_pred = lasso.predict(X_train)

# Calculate RMSE for the training set
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
print("Training RMSE: ", rmse_train)


# Make predictions and evaluate the model
y_pred = lasso.predict(X_val)
mse = mean_squared_error(y_val, y_pred)
rmse = np.sqrt(mse)
print("Validation RMSE: ", rmse)

# Inspect the coefficients
# print("Lasso coefficients: ", lasso.coef_)

#plot coeffs
feature_names=X_train.columns
coefficients=lasso.coef_
tmp_plot=pd.DataFrame([feature_names,coefficients])
# Plot the coefficients
plt.figure(figsize=(10, 6))
plt.barh(feature_names, coefficients, color='b')
plt.xlabel('Coefficient Value')
plt.title('Lasso Regression Coefficients')
plt.grid(True)
plt.show()
```

```python
# Create a DataFrame for better handling
coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
})

# Drop coefficients close to zero
coef_df = coef_df[np.abs(coef_df['Coefficient']) > 0.01]

# Sort the coefficients in descending order
coef_df = coef_df.sort_values(by='Coefficient', ascending=False)

# Plot the coefficients
plt.figure(figsize=(10, 8))
plt.barh(coef_df['Feature'], coef_df['Coefficient'], color='b')
plt.xlabel('Coefficient Value')
plt.title('Lasso Regression Coefficients')
plt.gca().invert_yaxis()  # Highest values on top
plt.grid(True)

# Make the y-axis labels more readable
plt.yticks(ticks=range(len(coef_df)), labels=coef_df['Feature'], rotation=0, ha='right')

plt.show()
```

### optimize lasso

```python
from sklearn.linear_model import LassoCV
# Define the model with LassoCV
lasso_cv = LassoCV(alphas=np.logspace(-4, 4, 50), cv=5, random_state=42)

# Train the model with cross-validation
lasso_cv.fit(X_train, y_train)

# Get the best alpha value
best_alpha = lasso_cv.alpha_
print("Best alpha: ", best_alpha)

# Make predictions on the training set
y_train_pred = lasso_cv.predict(X_train)

# Calculate RMSE for the training set
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
print("Training RMSE: ", rmse_train)

# Make predictions and evaluate the model
y_pred_cv = lasso_cv.predict(X_val)
mse_cv = mean_squared_error(y_val, y_pred_cv)
rmse_cv = np.sqrt(mse_cv)
print("Validation RMSE with best alpha: ", rmse_cv)
```
